Определим функции-правила, по которым определяется подозрительный номер. Каждая функция-правило возвращает либо 1, либо 0. Результаты функции суммируются и, если сумма превышает определенный порог, то номер блокируется.

##### Правило 1 - первая цафра номера 9
##### Правило 2 - вторая цифра номер > 7
##### Правило 3 - 3-я цифра номера равна 2
##### Правило 4 - 7-я цифра номера меньше 12-той 
##### Правило 5 - количество вызовов с этого номера больше 6 (определено кластризацией)

Реализуем ранжирующую систему


```python
def rule_1(separate_number):
    return 1 if(separate_number[0] == 9) else 0

def rule_2(separate_number):
    return 1 if(separate_number[2] == 2) else 0

def rule_3(separate_number):
     return 1 if(separate_number[1] >= 7) else 0

def rule_4(separate_number):
    return 1 if(separate_number[6] <= separate_number[11]) else 0

def rule_5(numb_call):
    return 1 if(numb_call >= 6) else 0 
    
```

Проверим правила

```python
rules_number = [rule_1, rule_2, rule_3, rule_4]
rules_call = [rule_5]
```

```python
limit_r = 4
```

```python
filtered_numbers=pd.DataFrame(columns=['NUMBER',  'CALL_CNT', 'score'])
```

```python
j=0
for i in range(0, len(data)):
    digits = split_phone_number_to_columns(data['ANUMBER_NORM'][i]) 
    sum = 0
    for a in rules_number:
        sum += a(digits)
    for a in rules_call:
        sum += a(data['CALL_CNT'][i])

    if sum >= limit_r:
        filtered_numbers.loc[j] = [data['ANUMBER_NORM'][i], data['CALL_CNT'][i], sum]
        j+=1
    
```

```python
filtered_numbers
```

Реализуем систему на паттернах

```python
patterns = [
            r'^3[7-8][0][9][0-9]{8}$',
            r'^3[7-8][5][2-4][0-9]{8}$',
            r'^4[4][1][0-4][0-9]{8}$',
            r'^9[7-9][0-5][0][0-9]{8}$',
            r'^9[7-9][0-5][8-9][0-9]{8}$',
            r'^9[7-9][0-5][1-2][0-9]{8}$',
            r'^9[7-9][0-5][4-5][0-9]{8}$',
    
           ]
```

```python
def is_flash_call_number(number):
   for pattern in patterns:
       if re.match(pattern, number):
           return True
   return False
```

```python
filtered_numbers_for_pattern=pd.DataFrame(columns=['NUMBER',  'CALL_CNT'])
```

```python
j = 0
for i in range(0, len(data)):
    number = str(data['ANUMBER_NORM'][i])
    #print(f'{number}: {is_flash_call_number(number)}')
    if is_flash_call_number(number):
        filtered_numbers_for_pattern.loc[j] = [data['ANUMBER_NORM'][i], data['CALL_CNT'][i]]
        j+=1
```

Объеденим два метода 

```python
s_numbs = pd.merge(filtered_numbers, filtered_numbers_for_pattern, on='NUMBER', how='inner')
s_numbs
```
